```python
from Mental_Health.utils.main_utils import read_yaml_file
schema_config = read_yaml_file(file_path=r'../config/schema.yaml')
```

```python
data_null_handler=schema_config
```

```python
# Replace 'UNKNOWN' with the specified values from the YAML file
def replace_unknown_values(df: pd.DataFrame, data_null_handler: dict) -> pd.DataFrame:
    for column, replacement in data_null_handler['data_null_handler'].items():
        if column in df.columns:
            df[column] = df[column].replace('Unknown', replacement[0])
    return df
```

```python
# Apply the function to the DataFrame
df = replace_unknown_values(df, data_null_handler)
```
